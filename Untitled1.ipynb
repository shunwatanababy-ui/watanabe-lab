```python
from pyscf import gto, dft
from pyscf.geomopt import geometric_solver

# 1. 分子の定義（ラジカルカチオン：電荷+1, スピン多重度2）
mol = gto.M(
    atom='''
        S  -1.50   0.65   0.00
        C  -2.35  -0.10   0.00
        C  -1.75  -1.35   0.00
        C  -0.50  -1.35   0.00
        C  -0.30  -0.10   0.00
        S   1.50  -0.65   0.00
        C   2.35   0.10   0.00
        C   1.75   1.35   0.00
        C   0.50   1.35   0.00
        C   0.30   0.10   0.00
    ''',
    charge=1,      # 電荷 +1 (電子を1個抜いたラジカルカチオン)
    spin=1,        # 不対電子1個 (2S+1 = 2)
    basis='6-31G*' # 標準的な基底関数
)

# 2. 開殻系のKohn-Sham DFT計算の設定 (B3LYP)
mf = dft.UKS(mol)
mf.xc = 'B3LYP'

# 3. 構造最適化（Geometry Optimization）の実行
print("--- DFT構造最適化を開始します（少し時間がかかります） ---")
mol_eq = geometric_solver.optimize(mf)
print("--- 構造最適化が完了しました！ ---")
```
