```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

def calculate_and_plot_pbttt(n_units=4, epsilon_T=0.0, epsilon_TT=-0.15, t_T_TT=-1.2, t_T_T=-1.0):
    """
    pBTTTオリゴマーの電子状態をTight-Binding近似により計算・プロットする関数
    構造単位: - T - TT - T - (3サイト/モノマー)
    """
    n_sites = 3 * n_units
    H = np.zeros((n_sites, n_sites))
    
    # --- ハミルトニアンの構築 ---
    for i in range(n_sites):
        # 1. オンサイトエネルギー（対角成分）
        if i % 3 == 1:
            H[i, i] = epsilon_TT  # チエノチオフェン(TT)環
        else:
            H[i, i] = epsilon_T   # チオフェン(T)環
            
        # 2. ホッピング（非対角成分）
        if i < n_sites - 1:
            if i % 3 == 0 or i % 3 == 1:
                H[i, i+1] = t_T_TT
                H[i+1, i] = t_T_TT
            elif i % 3 == 2:
                H[i, i+1] = t_T_T
                H[i+1, i] = t_T_T

    # 固有値と固有ベクトルの計算
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    
    # --- グラフの描画 ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # 左: エネルギー準位
    for idx, E in enumerate(eigenvalues):
        ax1.axhline(y=E, color='royalblue', linestyle='-', linewidth=2, alpha=0.7)
        if n_units <= 6:  # 単位数が少ない場合のみ準位の値をテキスト表示
            ax1.text(0.05, E + 0.03, f"State {idx+1}: {E:.3f} eV", fontsize=9, color='darkblue')
            
    ax1.set_title(f"Energy Levels (N = {n_units} Monomers)", fontsize=12, fontweight='bold')
    ax1.set_ylabel("Energy (eV)", fontsize=11)
    ax1.set_xlim(-0.1, 1.1)
    ax1.set_xticks([])
    ax1.grid(axis='y', linestyle=':', alpha=0.6)
    
    # 右: 最高エネルギー状態の存在確率密度 (|ψ|^2)
    target_state_idx = -1  # 最上位状態
    wave_amplitude = eigenvectors[:, target_state_idx]
    probability_density = wave_amplitude ** 2
    
    sites = np.arange(1, n_sites + 1)
    site_colors = ['#ff7f0e' if i % 3 == 1 else '#1f77b4' for i in range(n_sites)]
    
    ax2.bar(sites, probability_density, color=site_colors, alpha=0.85, edgecolor='black')
    
    site_labels = ["TT" if i % 3 == 1 else "T" for i in range(n_sites)]
    ax2.set_xticks(sites)
    ax2.set_xticklabels(site_labels, fontsize=9)
    ax2.set_title(f"Probability Density ($|\psi|^2$) of Highest State", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Backbone Site Sequence", fontsize=11)
    ax2.set_ylabel("Probability Density", fontsize=11)
    ax2.grid(axis='y', linestyle='--', alpha=0.5)
    
    # 凡例
    legend_elements = [Patch(facecolor='#1f77b4', edgecolor='black', label='Thiophene (T)'),
                       Patch(facecolor='#ff7f0e', edgecolor='black', label='Thienothiophene (TT)')]
    ax2.legend(handles=legend_elements, loc='upper right')
    
    plt.tight_layout()
    plt.show()

print("関数 'calculate_and_plot_pbttt' が正常にロードされました。")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def build_and_solve_hamiltonian(n_units, epsilon_T, epsilon_TT, t_T_TT, t_T_T):
    """
    指定されたパラメータからハミルトニアン行列を構築し、固有値（準位）を返すヘルパー関数
    """
    n_sites = 3 * n_units
    H = np.zeros((n_sites, n_sites))
    
    for i in range(n_sites):
        # オンサイトエネルギーの設定
        if i % 3 == 1:
            H[i, i] = epsilon_TT  # TT環
        else:
            H[i, i] = epsilon_T   # T環
            
        # ホッピング積分の設定
        if i < n_sites - 1:
            if i % 3 == 0 or i % 3 == 1:
                H[i, i+1] = t_T_TT
                H[i+1, i] = t_T_TT
            elif i % 3 == 2:
                H[i, i+1] = t_T_T
                H[i+1, i] = t_T_T
                
    eigenvalues = np.linalg.eigvalsh(H)
    return eigenvalues

def calculate_homo_lumo_evolution(max_units=10):
    """
    モノマー数 N = 1 から max_units までの HOMO と LUMO の変化を計算・プロットする
    """
    # --- pBTTTのタイトバインディング係数の仮設定 (eV単位) ---
    # HOMO計算用（価電子帯側）
    homo_params = {
        'epsilon_T': -5.1,   # 孤立チオフェン環のHOMO推定値
        'epsilon_TT': -4.9,  # 孤立チエノチオフェン環のHOMO（一般にTより少し浅い）
        't_T_TT': -0.7,      # 単位内結合
        't_T_T': -0.5        # 単位間結合
    }
    
    # LUMO計算用（伝導帯側）
    lumo_params = {
        'epsilon_T': -1.8,   # 孤立チオフェン環のLUMO推定値
        'epsilon_TT': -2.1,  # 孤立チエノチオフェン環のLUMO（一般にTより深い）
        't_T_TT': -0.6,      # 単位内結合
        't_T_T': -0.4        # 単位間結合
    }
    
    n_range = np.arange(1, max_units + 1)
    homo_levels = []
    lumo_levels = []
    gaps = []
    
    for n in n_range:
        # 1. HOMO側の全準位を計算し、最高エネルギーを取り出す
        # (タイトバインディングの符号の定義上、最上位を取り出す)
        homo_all = build_and_solve_hamiltonian(n, **homo_params)
        homo = np.max(homo_all)
        homo_levels.append(homo)
        
        # 2. LUMO側の全準位を計算し、最低エネルギーを取り出す
        # (LUMOバンドの最下位を取り出す)
        lumo_all = build_and_solve_hamiltonian(n, **lumo_params)
        lumo = np.min(lumo_all)
        lumo_levels.append(lumo)
        
        # 3. HOMO-LUMOギャップ (Eg)
        gaps.append(lumo - homo)
        
    # --- グラフ描画 ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # 左グラフ: HOMO / LUMO レベルのオリゴマー長依存性
    ax1.plot(n_range, lumo_levels, 'o-', color='crimson', linewidth=2, label='LUMO')
    ax1.plot(n_range, homo_levels, 'o-', color='navy', linewidth=2, label='HOMO')
    
    # 各点にエネルギー値をテキスト表示 (見やすさのため一部)
    for i, n in enumerate(n_range):
        if n % 2 == 1 or n == max_units:
            ax1.text(n, lumo_levels[i] + 0.08, f"{lumo_levels[i]:.2f} eV", ha='center', fontsize=9, color='darkred')
            ax1.text(n, homo_levels[i] - 0.15, f"{homo_levels[i]:.2f} eV", ha='center', fontsize=9, color='darkblue')
            
    ax1.set_title("HOMO and LUMO Levels vs Oligomer Length", fontsize=12, fontweight='bold')
    ax1.set_xlabel("Number of Monomer Units (N)", fontsize=11)
    ax1.set_ylabel("Energy (eV)", fontsize=11)
    ax1.set_xticks(n_range)
    ax1.set_ylim(-6.5, -1.0)
    ax1.grid(True, linestyle=':', alpha=0.6)
    ax1.legend(loc='center right', fontsize=11)
    
    # 右グラフ: HOMO-LUMOギャップの変化
    ax2.plot(n_range, gaps, 's--', color='purple', linewidth=2, label='Band Gap ($E_g$)')
    for i, n in enumerate(n_range):
        if n % 2 == 1 or n == max_units:
            ax2.text(n, gaps[i] + 0.05, f"{gaps[i]:.2f} eV", ha='center', fontsize=9, color='purple')
            
    ax2.set_title("HOMO-LUMO Gap Convergence", fontsize=12, fontweight='bold')
    ax2.set_xlabel("Number of Monomer Units (N)", fontsize=11)
    ax2.set_ylabel("Energy Gap (eV)", fontsize=11)
    ax2.set_xticks(n_range)
    ax2.set_ylim(min(gaps) - 0.3, max(gaps) + 0.3)
    ax2.grid(True, linestyle=':', alpha=0.6)
    ax2.legend(loc='upper right', fontsize=11)
    
    plt.tight_layout()
    plt.show()

# 計算の実行（モノマー数1から10まで）
calculate_homo_lumo_evolution(max_units=10)
```

```python
import numpy as np

def calculate_p3ht_monomer_g_tensor():
    """
    P3HTモノマー（チオフェン環）を模した簡易的な3軌道（px, py, pz）Tight-Binding模型から
    摂動論を用いてESRのgテンソル（g-shift）を近似計算するコード
    """
    # 自由電子のg値
    g_e = 2.002319
    
    # 1. 簡易的なフロンティア軌道（HOMO周辺）のエネルギー準位 (eV)
    # チオフェンのHOMOは主にπ軌道(pz)で構成される
    # 励起状態として、σ軌道や硫黄の孤立電子対（px, py）を想定
    E_homo = -5.1   # HOMO (pzベース)
    E_lumo = -1.8   # LUMO (pz*ベース)
    E_sigma1 = -6.5 # 深いσ軌道 (px, py混合)
    E_sigma2 = -7.0 # さらに深いσ軌道
    
    # 軌道のリスト [エネルギー, 軌道の主成分表現(x, y, z)]
    # 簡易的に各状態における px, py, pz の存在確率振幅（係数）を定義
    states = [
        {"name": "σ2",   "E": E_sigma2, "coeff": np.array([0.7, 0.7, 0.0])},
        {"name": "σ1",   "E": E_sigma1, "coeff": np.array([0.7, -0.7, 0.0])},
        {"name": "HOMO", "E": E_homo,   "coeff": np.array([0.0, 0.0, 1.0])}, # π軌道(z軸垂直)
        {"name": "LUMO", "E": E_lumo,   "coeff": np.array([0.0, 0.0, 1.0])}  # π*軌道
    ]
    
    # HOMOのインデックス（ターゲット状態）
    homo_idx = 2
    psi_n = states[homo_idx]
    
    # 2. スピン軌道相互作用係数 (硫黄原子 S の SOC定数 ξ は比較的大きい ~ 0.04 eV)
    xi_sulfur = 0.04 
    
    # 3. 軌道角運動量演算子 L_x, L_y, L_z の行列要素（p軌道基底 [px, py, pz] 内での作用）
    # L_x は py と pz を混ぜる: L_x |py> = i|pz>, L_x |pz> = -i|py>
    # L_y は pz と px を混ぜる: L_y |pz> = i|px>, L_y |px> = -i|pz>
    # L_z は px と py を混ぜる: L_z |px> = i|py>, L_z |py> = -i|px>
    
    # 各軸のgシフト初期化
    delta_g = np.zeros(3) # [Δg_xx, Δg_yy, Δg_zz]
    
    # 二次摂動の計算: sum_{m != n} 2 * <n|ξ L_μ|m><m|L_μ|n> / (E_n - E_m)
    # ※有機分子の通常の符号定義と、結合軌道・反結合軌道への寄与を模した簡易記述
    for m_idx, psi_m in enumerate(states):
        if m_idx == homo_idx:
            continue
            
        dE = psi_n["E"] - psi_m["E"]
        if dE == 0:
            continue
            
        # --- x方向の主軸シフト (L_x は y と z の結合を制御) ---
        # HOMO(z) と 別の軌道(y成分) の重なりを評価
        overlap_x = psi_n["coeff"][2] * psi_m["coeff"][1]
        # 摂動項への寄与 (因子 2 はスピン軌道項の定数由来)
        delta_g[0] += (2.0 * xi_sulfur * (overlap_x ** 2)) / dE
        
        # --- y方向の主軸シフト (L_y は z と x の結合を制御) ---
        # HOMO(z) と 別の軌道(x成分) の重なりを評価
        overlap_y = psi_n["coeff"][2] * psi_m["coeff"][0]
        delta_g[1] += (2.0 * xi_sulfur * (overlap_y ** 2)) / dE
        
        # --- z方向の主軸シフト (L_z は x と y の結合を制御) ---
        # HOMOはz成分のみなので、L_zはHOMOを動かさず、面内σ軌道同士を混ぜるためHOMOへの寄与は極めて小さい
        delta_g[2] += 0.0 

    # 4. 最終的なg値の算出 (g = g_e + Δg)
    g_tensor_diagonal = g_e + delta_g
    
    # 結果の表示
    print("==============================================")
    print("   P3HTモノマーの簡易 gテンソル計算結果 (主軸成分)")
    print("==============================================")
    print(f"自由電子のg値 (g_e): {g_e:.6f}\n")
    print(f"g_xx (主鎖・面内横方向): {g_tensor_diagonal[0]:.6f}  (Δg_xx: {delta_g[0]:.6g})")
    print(f"g_yy (主鎖・分子長軸方向): {g_tensor_diagonal[1]:.6f}  (Δg_yy: {delta_g[1]:.6g})")
    print(f"g_zz (分子面に垂直方向): {g_tensor_diagonal[2]:.6f}  (Δg_zz: {delta_g[2]:.6g})")
    print("----------------------------------------------")
    print(f"平均 g値 (g_iso): {np.mean(g_tensor_diagonal):.6f}")
    print("==============================================")

# 関数の実行
calculate_p3ht_monomer_g_tensor()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_spin_density_with_polarization(n_units=8, lambda_coupling=0.4, max_iter=30):
    """
    pBTTTオリゴマー上に1つのスピン（ホール）が注入された際のスピン密度分布を計算する。
    ポラロン効果（スピン密度に比例して格子・サイトエネルギーが安定化する効果）を考慮。
    
    Parameters:
    -----------
    n_units : int
        モノマー単位数 (1単位=3サイト, 計3*Nサイト)
    lambda_coupling : float
        電子-格子相互作用の強さ (0にすると完全に非局在化したスピン密度になる)
    max_iter : int
        自己収束計算の最大ループ回数
    """
    n_sites = 3 * n_units
    
    # 1. 基本パラメータの設定 (HOMO側)
    eps_T_base = -5.1
    eps_TT_base = -4.9
    t_T_TT = -0.7
    t_T_T = -0.5
    
    # 初期状態：スピン密度分布（確率密度）を均等に割り振る
    spin_density = np.ones(n_sites) / n_sites
    
    # 収束計算ループ
    for iteration in range(max_iter):
        old_spin_density = spin_density.copy()
        
        # ハミルトニアンの再構築
        H = np.zeros((n_sites, n_sites))
        for i in range(n_sites):
            # オンサイトエネルギー (ポラロン効果により、スピン密度が大きいサイトほどエネルギーが浅く/深くなる)
            # ここでは結合エネルギーの局所変化として、スピン密度に比例してポテンシャルを変化させる
            if i % 3 == 1:
                H[i, i] = eps_TT_base + lambda_coupling * spin_density[i]
            else:
                H[i, i] = eps_T_base + lambda_coupling * spin_density[i]
                
            # ホッピング成分
            if i < n_sites - 1:
                if i % 3 == 0 or i % 3 == 1:
                    H[i, i+1] = t_T_TT
                    H[i+1, i] = t_T_TT
                elif i % 3 == 2:
                    H[i, i+1] = t_T_T
                    H[i+1, i] = t_T_T
                    
        # 対角化して固有値と固有ベクトルを取得
        eigenvalues, eigenvectors = np.linalg.eigh(H)
        
        # タイトバインディングの符号定義上、最高エネルギー状態（index=-1）がHOMO軌道
        homo_vector = eigenvectors[:, -1]
        
        # 新しいスピン密度（確率密度 |\psi|^2）を計算
        spin_density = homo_vector ** 2
        
        # 収束判定 (変化が極小になったらストップ)
        if np.linalg.norm(spin_density - old_spin_density) < 1e-5:
            break

    # --- プロット処理 ---
    plt.figure(figsize=(10, 5))
    sites = np.arange(1, n_sites + 1)
    
    # サイトごとの色分け (Tは青、TTはオレンジ)
    colors = ['#1f77b4' if i % 3 != 1 else '#ff7f0e' for i in range(n_sites)]
    
    plt.bar(sites, spin_density, color=colors, alpha=0.8, edgecolor='black', width=0.8)
    plt.plot(sites, spin_density, color='gray', linestyle='--', alpha=0.5)
    
    # 軸の設定
    labels = ["TT" if i % 3 == 1 else "T" for i in range(n_sites)]
    plt.xticks(sites, labels, fontsize=9)
    plt.xlabel("Polymer Backbone Site Sequence", fontsize=12)
    plt.ylabel("Spin Density ($|\psi_{HOMO}|^2$)", fontsize=12)
    plt.title(f"Spin Density Distribution on pBTTT Oligomer ($N$={n_units} monomers)\nWith Polaron Relaxation Effect", fontsize=13, fontweight='bold')
    
    # 凡例表示
    from matplotlib.patches import Patch
    legend_elements = [Patch(facecolor='#1f77b4', edgecolor='black', label='Thiophene (T)'),
                       Patch(facecolor='#ff7f0e', edgecolor='black', label='Thienothiophene (TT)')]
    plt.legend(handles=legend_elements, loc='upper right')
    
    plt.grid(axis='y', linestyle=':', alpha=0.6)
    plt.tight_layout()
    plt.show()
    
    # 各サイトのスピン密度をコンソールに出力
    print(f"総スピン量（合計値の検証）: {np.sum(spin_density):.4f} (電子1個分)")
    print(f"中央部の最大スピン密度: {np.max(spin_density):.4f}")

# 8量体（24サイト）でスピン密度を計算
calculate_spin_density_with_polarization(n_units=8, lambda_coupling=0.6)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_p3ht_spin_density(n_units=12, lambda_coupling=0.5, disorder_sigma=0.0, max_iter=40):
    """
    P3HTオリゴマー上に1つのスピン（ホール）が注入された際のスピン密度分布を計算する。
    1単位(Monomer) = 1つのチオフェン環 (計 N サイトの単純な鎖状モデル)
    
    Parameters:
    -----------
    n_units : int
        チオフェン環の数 (オリゴマー長 N)
    lambda_coupling : float
        電子-格子相互作用（ポラロン効果）の強さ。
        値を大きくするほど、スピン自身が作る格子の歪みにトラップされて中央に局在化します。
    disorder_sigma : float
        主鎖のねじれ等を模した、ホッピング積分のランダムな揺らぎ（標準偏差 eV）。
    max_iter : int
        自己収束計算の最大ループ回数
    """
    n_sites = n_units
    
    # 1. 基本パラメータの設定 (P3HTのHOMO側)
    eps_T_base = -5.1  # 孤立チオフェン環の基本オンサイトエネルギー (eV)
    t_base = -0.6      # 隣接するチオフェン環同士の基本ホッピング積分 (eV)
    
    # 構造無秩序（ねじれ等）によるホッピング積分の揺らぎを事前に生成
    np.random.seed(42) # 再現性のためシードを固定
    t_disorder = np.random.normal(0, disorder_sigma, n_sites - 1)
    
    # 初期状態：スピン密度分布（確率密度）を全サイトに均等に割り振る
    spin_density = np.ones(n_sites) / n_sites
    
    # 収束計算（Self-Consistent Loop）
    for iteration in range(max_iter):
        old_spin_density = spin_density.copy()
        
        # ハミルトニアン行列の初期化
        H = np.zeros((n_sites, n_sites))
        
        for i in range(n_sites):
            # オンサイトエネルギー: 
            # ポラロン緩和により、スピン（ホール）密度が大きいサイトほど、
            # 局所的な格子歪みによってエネルギー的に自己安定化（深く）する
            H[i, i] = eps_T_base + lambda_coupling * spin_density[i]
            
            # ホッピング成分（隣り合うチオフェン環同士の結合）
            if i < n_sites - 1:
                t_actual = t_base + t_disorder[i]
                H[i, i+1] = t_actual
                H[i+1, i] = t_actual
                
        # 行列の対角化
        eigenvalues, eigenvectors = np.linalg.eigh(H)
        
        # ホール（スピン）が占有する最高占有軌道（HOMO）の波動関数を取り出す
        # タイトバインディングの定義上、固有値が最も高い状態(index=-1)がHOMOに対応
        homo_vector = eigenvectors[:, -1]
        
        # 波動関数の二乗から、新しいスピン密度（確率密度）を計算
        spin_density = homo_vector ** 2
        
        # 収束判定 (前ステップからの変化が極小になったら終了)
        if np.linalg.norm(spin_density - old_spin_density) < 1e-6:
            break

    # --- グラフ描画 ---
    plt.figure(figsize=(9, 4.5))
    sites = np.arange(1, n_sites + 1)
    
    # 棒グラフとトレンド線のプロット
    plt.bar(sites, spin_density, color='crimson', alpha=0.75, edgecolor='black', width=0.7, label='Spin Density')
    plt.plot(sites, spin_density, color='darkred', linestyle='--', alpha=0.6)
    
    # 軸・ラベルの設定
    plt.xlabel("Thiophene Ring Position along Backbone", fontsize=11)
    plt.ylabel(r"Spin Density ($|\psi_{HOMO}|^2$)", fontsize=11)
    plt.title(r"Spin Density Distribution on P3HT Oligomer ($N$={})".format(n_units), fontsize=12, fontweight='bold')
    plt.xticks(sites)
    plt.grid(axis='y', linestyle=':', alpha=0.6)
    
    # パラメータ情報のテキスト表示
    info_text = f"Polaron Coupling ($\lambda$): {lambda_coupling} eV\nDisorder ($\sigma$): {disorder_sigma} eV"
    plt.gca().text(0.05, 0.92, info_text, transform=plt.gca().transAxes, 
                   fontsize=9, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    print(f"計算収束回数: {iteration+1} 回")
    print(f"全サイトのスピン密度合計: {np.sum(spin_density):.4f} (不対電子1個分に一致)")

# 計算の実行例 (12量体、ポラロン効果あり、無秩序なし)
calculate_p3ht_spin_density(n_units=12, lambda_coupling=0.5, disorder_sigma=0.0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_p3ht_spin_density(n_units=12, lambda_coupling=0.5, disorder_sigma=0.0, max_iter=40):
    """
    P3HTオリゴマー上に1つのスピン（ホール）が注入された際のスピン密度分布を計算する。
    1単位(Monomer) = 1つのチオフェン環 (計 N サイトの単純な鎖状モデル)
    
    Parameters:
    -----------
    n_units : int
        チオフェン環の数 (オリゴマー長 N)
    lambda_coupling : float
        電子-格子相互作用（ポラロン効果）の強さ。
    disorder_sigma : float
        主鎖のねじれ等を模した、ホッピング積分のランダムな揺らぎ（標準偏差 eV）。
    max_iter : int
        自己収束計算の最大ループ回数
    """
    n_sites = n_units
    
    # 1. 基本パラメータの設定 (P3HTのHOMO側)
    eps_T_base = -5.1  # 孤立チオフェン環の基本オンサイトエネルギー (eV)
    t_base = -0.6      # 隣接するチオフェン環同士の基本ホッピング積分 (eV)
    
    # 構造無秩序（ねじれ等）によるホッピング積分の揺らぎを事前に生成
    np.random.seed(42) # 再現性のためシードを固定
    t_disorder = np.random.normal(0, disorder_sigma, n_sites - 1)
    
    # 初期状態：スピン密度分布（確率密度）を全サイトに均等に割り振る
    spin_density = np.ones(n_sites) / n_sites
    
    # 収束計算（Self-Consistent Loop）
    for iteration in range(max_iter):
        old_spin_density = spin_density.copy()
        
        # ハミルトニアン行列の初期化
        H = np.zeros((n_sites, n_sites))
        
        for i in range(n_sites):
            # オンサイトエネルギー: ポラロン緩和
            H[i, i] = eps_T_base + lambda_coupling * spin_density[i]
            
            # ホッピング成分（隣り合うチオフェン環同士の結合）
            if i < n_sites - 1:
                t_actual = t_base + t_disorder[i]
                H[i, i+1] = t_actual
                H[i+1, i] = t_actual
                
        # 行列の対角化
        eigenvalues, eigenvectors = np.linalg.eigh(H)
        
        # HOMO軌道の波動関数を取り出す
        homo_vector = eigenvectors[:, -1]
        
        # スピン密度（確率密度）を計算
        spin_density = homo_vector ** 2
        
        # 収束判定
        if np.linalg.norm(spin_density - old_spin_density) < 1e-6:
            break

    # --- グラフ描画 ---
    plt.figure(figsize=(9, 4.5))
    sites = np.arange(1, n_sites + 1)
    
    # 棒グラフとトレンド線のプロット
    plt.bar(sites, spin_density, color='crimson', alpha=0.75, edgecolor='black', width=0.7, label='Spin Density')
    plt.plot(sites, spin_density, color='darkred', linestyle='--', alpha=0.6)
    
    # 軸・ラベルの設定
    plt.xlabel("Thiophene Ring Position along Backbone", fontsize=11)
    plt.ylabel(r"Spin Density ($|\psi_{HOMO}|^2$)", fontsize=11)
    plt.title(r"Spin Density Distribution on P3HT Oligomer ($N$={})".format(n_units), fontsize=12, fontweight='bold')
    plt.xticks(sites)
    plt.grid(axis='y', linestyle=':', alpha=0.6)
    
    # ★ 修正箇所: 文字列の前に fr を付けて \lambda と \sigma のエスケープ警告を回避
    info_text = fr"Polaron Coupling ($\lambda$): {lambda_coupling} eV" + "\n" + fr"Disorder ($\sigma$): {disorder_sigma} eV"
    plt.gca().text(0.05, 0.92, info_text, transform=plt.gca().transAxes, 
                   fontsize=9, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    print(f"計算収束回数: {iteration+1} 回")
    print(f"全サイトのスピン密度合計: {np.sum(spin_density):.4f} (不対電子1個分に一致)")

# 計算の実行
calculate_p3ht_spin_density(n_units=12, lambda_coupling=0.5, disorder_sigma=0.0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_p3ht_atomic_spin_density():
    """
    P3HTモノマー（チオフェン環）を構成する各原子（C, S）の上での
    スピン密度分布（HOMOの確率密度）をタイトバインディング模型で計算する。
    
    原子のインデックス定義:
    0: S (硫黄)
    1: C_alpha1 (Sに隣接する炭素)
    2: C_beta1  (中央寄りの炭素)
    3: C_beta2  (中央寄りの炭素)
    4: C_alpha2 (Sに隣接する炭素)
    """
    n_atoms = 5
    H = np.zeros((n_atoms, n_atoms))
    
    # 1. オンサイトエネルギー（各原子の基本エネルギーレベル, eV）
    # 硫黄(S)は炭素(C)よりも電気陰性度が高いため、軌道エネルギーが深くなります。
    alpha_C = 0.0      # 炭素の基準値
    alpha_S = -1.2     # 硫黄のエネルギー（Cより深い）
    
    H[0, 0] = alpha_S
    H[1, 1] = alpha_C
    H[2, 2] = alpha_C
    H[3, 3] = alpha_C
    H[4, 4] = alpha_C
    
    # 2. ホッピング積分（隣接原子間の結合強さ, eV）
    # 炭素-炭素(C-C)結合と、炭素-硫黄(C-S)結合で異なる値を設定します。
    beta_CC = -2.5     # C-C 結合
    beta_CS = -1.8     # C-S 結合（一般にC-Cより弱い）
    
    # S(0) - C_alpha1(1)
    H[0, 1] = H[1, 0] = beta_CS
    # C_alpha1(1) - C_beta1(2)
    H[1, 2] = H[2, 1] = beta_CC
    # C_beta1(2) - C_beta2(3)
    H[2, 3] = H[3, 2] = beta_CC
    # C_beta2(3) - C_alpha2(4)
    H[3, 4] = H[4, 3] = beta_CC
    # C_alpha2(4) - S(0)  (環を閉じる)
    H[4, 0] = H[0, 4] = beta_CS
    
    # 3. ハミルトニアンの対角化
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    
    # HOMO（最高占有軌道）の取り出し
    # タイトバインディングの定義上、不対電子（ホール）が入るエネルギーが最も高い軌道
    # (ここでは5つの軌道のうち、エネルギーが上から2番目か3番目など、電子数に応じて決まりますが、
    #  チオフェンラジカルカチオンの不対電子軌道に対応する、全電子数の「最上位占有軌道」を選択)
    # 4電子ラジカルを想定した場合、上から2番目の軌道(index=3)がHOMOに相当します。
    homo_idx = 3 
    homo_vector = eigenvectors[:, homo_idx]
    
    # 各原子上のスピン密度（波動関数の存在確率振幅の二乗）
    spin_density = homo_vector ** 2
    
    # --- グラフの描画 ---
    plt.figure(figsize=(8, 4.5))
    atom_labels = [r'S (0)', r'$C_{\alpha 1}$ (1)', r'$C_{\beta 1}$ (2)', r'$C_{\beta 2}$ (3)', r'$C_{\alpha 2}$ (4)']
    x_pos = np.arange(n_atoms)
    
    # 原子種ごとに色を変える（Sはオレンジ、Cはグレー）
    colors = ['#ff7f0e', '#1f77b4', '#1f77b4', '#1f77b4', '#1f77b4']
    
    plt.bar(x_pos, spin_density, color=colors, alpha=0.8, edgecolor='black', width=0.6)
    
    # 各バーの上に数値をテキスト表示
    for i, val in enumerate(spin_density):
        plt.text(i, val + 0.01, f"{val:.3f}", ha='center', fontsize=10, fontweight='bold')
        
    plt.xticks(x_pos, atom_labels, fontsize=11)
    plt.ylabel(r"Spin Density ($|\psi_{HOMO}|^2$)", fontsize=12)
    plt.title("Atomic Spin Density Distribution on a P3HT Monomer Ring", fontsize=13, fontweight='bold')
    plt.ylim(0, max(spin_density) + 0.08)
    plt.grid(axis='y', linestyle=':', alpha=0.6)
    
    plt.tight_layout()
    plt.show()
    
    # 詳細な数値出力
    print("====================================")
    print("  各原子上のスピン密度（存在確率）")
    print("====================================")
    for label, dens in zip(atom_labels, spin_density):
        clean_label = label.replace('$', '').replace('\\alpha ', 'α').replace('\\beta ', 'β')
        print(f"{clean_label:<12} : {dens:.4f}")
    print("------------------------------------")
    print(f"総スピン量合計: {np.sum(spin_density):.4f} (電子1個分)")
    print("====================================")

# 計算の実行
calculate_p3ht_atomic_spin_density()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def calculate_and_plot_atomic_spin_colormap():
    """
    P3HTモノマー（チオフェン環）の原子レベルTight-Binding計算を行い、
    5員環の原子配置に合わせてスピン密度分布を2次元カラーマップとして描画する。
    """
    # --- 1. タイトバインディング計算 (前述の模型) ---
    n_atoms = 5
    H = np.zeros((n_atoms, n_atoms))
    
    alpha_C = 0.0      # 炭素のオンサイト
    alpha_S = -1.2     # 硫黄のオンサイト（Cより深い）
    beta_CC = -2.5     # C-C ホッピング
    beta_CS = -1.8     # C-S ホッピング
    
    H[0, 0] = alpha_S
    for i in range(1, 5):
        H[i, i] = alpha_C
        
    H[0, 1] = H[1, 0] = beta_CS
    H[1, 2] = H[2, 1] = beta_CC
    H[2, 3] = H[3, 2] = beta_CC
    H[3, 4] = H[4, 3] = beta_CC
    H[4, 0] = H[0, 4] = beta_CS
    
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 3  # 不対電子軌道 (HOMO)
    spin_density = eigenvectors[:, homo_idx] ** 2

    # --- 2. 2次元平面上へのチオフェン環の原子座標の定義 ---
    # 五角形の形状になるように近似的な (x, y) 座標を設定
    # 0: S, 1: C_alpha1, 2: C_beta1, 3: C_beta2, 4: C_alpha2
    coords = np.array([
        [0.0,  0.6],   # S (上部頂点)
        [-0.9, 0.0],   # C_alpha1 (左肩)
        [-0.6, -1.0],  # C_beta1  (左底)
        [0.6,  -1.0],  # C_beta2  (右底)
        [0.9,  0.0]    # C_alpha2 (右肩)
    ])
    
    atom_names = ['S', r'$C_{\alpha 1}$', r'$C_{\beta 1}$', r'$C_{\beta 2}$', r'$C_{\alpha 2}$']

    # --- 3. 描画処理 ---
    fig, ax = plt.subplots(figsize=(7, 6))
    
    # 結合（原子間の境界線）をプロット
    bond_pairs = [(0,1), (1,2), (2,3), (3,4), (4,0)]
    for p in bond_pairs:
        ax.plot(coords[p, 0], coords[p, 1], color='gray', linestyle='-', linewidth=2, zorder=1)
        
    # カラーマップの設定 (お好みに合わせて 'Reds', 'viridis', 'plasma' などに変更可能)
    cmap = cm.get_cmap('YlOrRd') 
    max_dens = np.max(spin_density)
    
    # 各原子をスピン密度に応じた大きさ・色でプロット
    scatter = ax.scatter(
        coords[:, 0], coords[:, 1], 
        s=1200,                         # 丸の大きさ
        c=spin_density,                 # カラーマップを適用する値
        cmap=cmap, 
        vmin=0, vmax=max_dens + 0.05,   # カラーバーの範囲
        edgecolors='black', linewidths=2,
        zorder=2
    )
    
    # 原子名とスピン密度の数値を各原子の上にテキスト表示
    for i, (x, y) in enumerate(coords):
        # 視認性を良くするため、カラーマップが濃い場合は文字を白、薄い場合は黒にする簡易判定
        text_color = 'white' if spin_density[i] > max_dens * 0.6 else 'black'
        
        # 原子名
        ax.text(x, y + 0.05, atom_names[i], ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
        # スピン密度の数値
        ax.text(x, y - 0.05, f"{spin_density[i]:.3f}", ha='center', va='top', fontsize=10, fontweight='bold', color=text_color)

    # カラーバーの追加
    cbar = plt.colorbar(scatter, ax=ax, orientation='horizontal', pad=0.1, shrink=0.8)
    cbar.set_label('Spin Density Value', fontsize=11)
    
    # グラフの見栄えの調整
    ax.set_title("2D Mapping of Atomic Spin Density\non a P3HT Monomer (Thiophene Ring)", fontsize=13, fontweight='bold', pad=15)
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.2)
    ax.set_aspect('equal') # 縦横比を1:1にする
    ax.axis('off')         # 枠線やx-y軸の目盛りを非表示にする
    
    plt.tight_layout()
    plt.show()

# 実行
calculate_and_plot_atomic_spin_colormap()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_p3hs_spin_colormap():
    """
    P3HSモノマー（セレンフェン環）の原子レベルTight-Binding計算を行い、
    5員環の原子配置に合わせてスピン密度分布を2次元カラーマップとして描画する。
    """
    # --- 1. タイトバインディング計算 (P3HS用にパラメータを調整) ---
    n_atoms = 5
    H = np.zeros((n_atoms, n_atoms))
    
    # 【パラメータの変更点】
    # SeはSより電気陰性度が低いため、オンサイトを -1.2 から -0.6 へ浅く設定
    # C-Seは結合長が長いため、ホッピング（結合強さ）を -1.8 から -1.4 へ弱く設定
    alpha_C = 0.0      # 炭素のオンサイト
    alpha_Se = -0.6    # セレンのオンサイト（Sの -1.2 より浅い）
    beta_CC = -2.5     # C-C ホッピング
    beta_CSe = -1.4    # C-Se ホッピング（Sの -1.8 より弱い）
    
    H[0, 0] = alpha_Se
    for i in range(1, 5):
        H[i, i] = alpha_C
        
    # 環状の結合関係を構築
    H[0, 1] = H[1, 0] = beta_CSe
    H[1, 2] = H[2, 1] = beta_CC
    H[2, 3] = H[3, 2] = beta_CC
    H[3, 4] = H[4, 3] = beta_CC
    H[4, 0] = H[0, 4] = beta_CSe
    
    # ハミルトニアンの対角化
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 3  # 不対電子軌道 (HOMO)
    spin_density = eigenvectors[:, homo_idx] ** 2

    # --- 2. 2次元平面上へのセレンフェン環の原子座標の定義 ---
    # セレンは原子半径が大きいため、結合長を考慮して少し上部（Sのとき0.6だった y座標）を広げに設定
    coords = np.array([
        [0.0,  0.8],   # Se (上部頂点、原子半径を考慮して少し高めに変更)
        [-0.9, 0.0],   # C_alpha1 (左肩)
        [-0.6, -1.0],  # C_beta1  (左底)
        [0.6,  -1.0],  # C_beta2  (右底)
        [0.9,  0.0]    # C_alpha2 (右肩)
    ])
    
    atom_names = ['Se', r'$C_{\alpha 1}$', r'$C_{\beta 1}$', r'$C_{\beta 2}$', r'$C_{\alpha 2}$']

    # --- 3. 描画処理 ---
    fig, ax = plt.subplots(figsize=(7, 6))
    
    # 結合（原子間の境界線）をプロット
    bond_pairs = [(0,1), (1,2), (2,3), (3,4), (4,0)]
    for p in bond_pairs:
        ax.plot(coords[p, 0], coords[p, 1], color='gray', linestyle='-', linewidth=2, zorder=1)
        
    # 最新のMatplotlib推奨記述を使用
    cmap = plt.get_cmap('YlOrRd') 
    max_dens = np.max(spin_density)
    
    # 各原子をスピン密度に応じた大きさ・色でプロット
    scatter = ax.scatter(
        coords[:, 0], coords[:, 1], 
        s=1200,                         # 丸の大きさ
        c=spin_density,                 # カラーマップを適用する値
        cmap=cmap, 
        vmin=0, vmax=max_dens + 0.05,   # カラーバーの範囲
        edgecolors='black', linewidths=2,
        zorder=2
    )
    
    # 原子名とスピン密度の数値を各原子の上にテキスト表示
    for i, (x, y) in enumerate(coords):
        text_color = 'white' if spin_density[i] > max_dens * 0.6 else 'black'
        
        # 原子名
        ax.text(x, y + 0.06, atom_names[i], ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')
        # スピン密度の数値
        ax.text(x, y - 0.05, f"{spin_density[i]:.3f}", ha='center', va='top', fontsize=10, fontweight='bold', color=text_color)

    # カラーバーの追加
    cbar = plt.colorbar(scatter, ax=ax, orientation='horizontal', pad=0.1, shrink=0.8)
    cbar.set_label('Spin Density Value', fontsize=11)
    
    # グラフの見栄えの調整
    ax.set_title("2D Mapping of Atomic Spin Density\non a P3HS Monomer (Selenophene Ring)", fontsize=13, fontweight='bold', pad=15)
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.4)
    ax.set_aspect('equal')
    ax.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    # 数値のコンソール出力
    print("====================================")
    print("  P3HS（セレン置換）各原子上のスピン密度")
    print("====================================")
    for name, dens in zip(['Se (0)', 'C_alpha1 (1)', 'C_beta1 (2)', 'C_beta2 (3)', 'C_alpha2 (4)'], spin_density):
        print(f"{name:<14} : {dens:.4f}")
    print("====================================")

# 実行
calculate_and_plot_p3hs_spin_colormap()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_detailed_g_tensor_p3ht():
    """
    P3HTモノマー（チオフェン環）の原子スピン密度から、
    摂動論（スピン軌道相互作用）を考慮して詳細なgテンソルを算出する
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. 前工程：原子レベルタイトバインディングによるスピン密度計算
    # ========================================================
    n_atoms = 5
    H = np.zeros((n_atoms, n_atoms))
    
    alpha_C = 0.0
    alpha_S = -1.2
    beta_CC = -2.5
    beta_CS = -1.8
    
    H[0, 0] = alpha_S
    for i in range(1, 5):
        H[i, i] = alpha_C
        
    H[0, 1] = H[1, 0] = beta_CS
    H[1, 2] = H[2, 1] = beta_CC
    H[2, 3] = H[3, 2] = beta_CC
    H[3, 4] = H[4, 3] = beta_CC
    H[4, 0] = H[0, 4] = beta_CS
    
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 3  # HOMO軌道
    
    # 各原子上のスピン密度（c_i^2）
    # 0: S, 1,4: C_alpha, 2,3: C_beta
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S = spin_densities[0]       # 硫黄上のスピン密度
    rho_C_alpha = (spin_densities[1] + spin_densities[4]) / 2.0
    rho_C_beta = (spin_densities[2] + spin_densities[3]) / 2.0

    # ========================================================
    # 2. 本工程：スピン密度をベースにしたgテンソルの異方性計算
    # ========================================================
    # 物理パラメータ (提示論文の知見に基づく代表値)
    xi_S = 0.046   # 硫黄の原子スピン軌道相互作用定数 (eV)
    xi_C = 0.0035  # 炭素の原子スピン軌道相互作用定数 (eV, Sに比べて無視できるほど小さい)
    
    # 励起状態（σ軌道や孤立電子対 n軌道）との平均エネルギーギャップ (eV)
    # HOMO（π）から見た、面内軌道（σ, n）の深さ
    dE_sigma = 2.2  
    
    # 各軸方向のエッジマトリックス要素（幾何学的な因子の近似）
    # 磁場の方向（x, y, z）によって、どの軌道（px, py, pz）が交わるかが決まる
    # 2.0 はスピン軌道項の公式に現れる係数
    factor_x = 2.0 * 0.8  # x軸（分子面内・短軸）: 硫黄のn_sigma孤立電子対との重なりが大きい
    factor_y = 2.0 * 0.4  # y軸（分子面内・長軸、主鎖方向）
    factor_z = 2.0 * 0.0  # z軸（分子面に垂直）: π-π* 間の遷移（Lz演算子ではHOMO(pz)からpzへは禁制のためほぼ0）

    # 各軸のgシフト（Δg = ρ * ξ * factor / dE）
    # 主にスピン軌道相互作用の大きい硫黄(S)からの寄与を計算
    delta_g_x = (rho_S * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S * xi_S * factor_z) / dE_sigma # 基本的にゼロ近傍
    
    # 炭素（C）からの微小な寄与を補正として加算
    delta_g_x += (np.sum(spin_densities[1:]) * xi_C * 0.2) / dE_sigma

    # 最終的なg値の組み立て
    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # --- グラフでの可視化 ---
    fig, ax = plt.subplots(figsize=(8, 4))
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    # 自由電子の参照線
    ax.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    # 各主軸方向のg値をプロット
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax.plot(val, idx, 'o', markersize=12, color=colors[idx], label=label)
        ax.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax.set_yticks(range(3))
    ax.set_yticklabels(['X軸成分', 'Y軸成分', 'Z軸成分'], fontsize=11)
    ax.set_xlabel('g-value', fontsize=12)
    ax.set_title('Anisotropy of ESR g-tensor Calculated from Atomic Spin Density', fontsize=13, fontweight='bold')
    ax.set_xlim(2.001, 2.005)
    ax.grid(axis='x', linestyle=':', alpha=0.6)
    ax.legend(loc='lower right')
    
    plt.tight_layout()
    plt.show()

    # 詳細ログの出力
    print("====================================================")
    print("  原子スピン密度連動型：P3HT gテンソル詳細計算結果")
    print("====================================================")
    print(f"【前段計算】硫黄(S)上のスピン密度 (ρ_S): {rho_S:.4f}")
    print(f"【前段計算】炭素(C)上の総スピン密度  : {1.0 - rho_S:.4f}\n")
    print(f"g_xx (分子面内・短軸方向) : {g_xx:.6f}  (Δg_xx: {delta_g_x:+.6f})")
    print(f"g_yy (分子面内・主鎖方向) : {g_yy:.6f}  (Δg_yy: {delta_g_y:+.6f})")
    print(f"g_zz (分子面に垂直方向)   : {g_zz:.6f}  (Δg_zz: {delta_g_z:+.6f})")
    print("----------------------------------------------------")
    print(f"等方性 g値 (g_iso)        : {g_iso:.6f}")
    print("====================================================")

# 実行
calculate_detailed_g_tensor_p3ht()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_pbttt_fixed_analysis():
    """
    PBTTTモノマー（14原子モデル）の原子スピン密度を計算し、
    正しい幾何学的トポロジーに修正した2Dカラーマップとgテンソルを表示する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. タイトバインディング計算 (14原子骨格)
    # ========================================================
    n_atoms = 14
    H = np.zeros((n_atoms, n_atoms))
    
    alpha_C = 0.0
    alpha_S_T = -1.2   
    alpha_S_TT = -1.4  
    
    # 正しい原子インデックス定義（左から右へ番号順に骨格を繋ぎます）
    # 左T環:  0, 1, 2(S), 3, 4
    # 中央TT: 5, 6(S), 7, 8, 9(S), 10
    # 右T環:  11, 12, 13(S), 14 ... のように対応
    S_indices_T = [2, 12]
    S_indices_TT = [6, 9]
    
    for i in range(n_atoms):
        if i in S_indices_T:
            H[i, i] = alpha_S_T
        elif i in S_indices_TT:
            H[i, i] = alpha_S_TT
        else:
            H[i, i] = alpha_C
            
    beta_CC = -2.5
    beta_CS = -1.8
    
    # 主鎖に沿った一連の結合 (0 -> 1 -> 2 -> ... -> 13)
    for i in range(n_atoms - 1):
        if i in S_indices_T or (i+1) in S_indices_T or i in S_indices_TT or (i+1) in S_indices_TT:
            t = beta_CS
        else:
            t = beta_CC
        H[i, i+1] = H[i+1, i] = t
        
    # 5員環をしっかりと閉じるためのトポロジー結合（ここがズレていました）
    H[0, 4] = H[4, 0] = beta_CC  # 左T環を閉じる
    H[5, 10] = H[10, 5] = beta_CC # 中央TT環の共有部を閉じる
    H[3, 5] = H[5, 3] = beta_CC  # 左T環と中央TT環の接続
    H[10, 11] = H[11, 10] = beta_CC # 中央TT環と右T環の接続
    H[11, 13] = H[13, 11] = beta_CS # 右T環を閉じる

    # 対角化とHOMOの抽出
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 10 
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[S_indices_T]) + np.sum(spin_densities[S_indices_TT])

    # ========================================================
    # 2. gテンソルの異方性計算
    # ========================================================
    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.4 
    
    factor_x = 2.0 * 0.85 
    factor_y = 2.0 * 0.35 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 3. 描画処理 (左右2画面)
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # --------------------------------------------------------
    # 左画面: 2Dスピン密度カラーマップ (座標をきれいに再配置)
    # --------------------------------------------------------
    # 0〜13の番号が左から右へ綺麗に整列する幾何学構造へ修正
    coords = np.array([
        [-2.7, -0.4], [-2.7,  0.4], [-2.0,  0.8], [-1.3,  0.4], [-1.3, -0.4], # 左T環 (0-4)
        [-0.7, -0.4], [-0.0, -0.8], [ 0.7, -0.4],                         # 中央TT環下部 (5-7)
        [ 0.7,  0.4], [ 0.0,  0.8], [-0.7,  0.4],                         # 中央TT環上部 (8-10)
        [ 1.3, -0.4], [ 1.3,  0.4], [ 2.0,  0.8]                          # 右T環の一部 (11-13)
    ])
    
    # 修正されたトポロジーに基づく正確な結合ペア
    correct_bonds = [
        (0,1), (1,2), (2,3), (3,4), (4,0),  # 左T環
        (3,5), (4,10),                      # TとTTの境界
        (5,6), (6,7), (7,8), (8,9), (9,10), (10,5), # 中央TT環
        (7,11), (8,12),                     # TTとTの境界
        (11,12), (12,13)                    # 右T環の一部
    ]
    
    for b in correct_bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='gray', linestyle='-', linewidth=2, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=600, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.5, zorder=2
    )
    
    # 硫黄原子(S)の位置にラベルを明示
    for idx in range(n_atoms):
        if idx in S_indices_T:
            ax1.text(coords[idx, 0], coords[idx, 1] + 0.22, r'$S_T$', ha='center', fontsize=12, fontweight='bold', color='darkred')
        elif idx in S_indices_TT:
            ax1.text(coords[idx, 0], coords[idx, 1] + 0.22, r'$S_{TT}$', ha='center', fontsize=12, fontweight='bold', color='darkblue')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=9, fontweight='bold', color=txt_color)

    ax1.set_title("Corrected 2D Mapping of Atomic Spin Density\non PBTTT Monomer (T-TT-T)", fontsize=12, fontweight='bold')
    ax1.set_xlim(-3.4, 2.7)
    ax1.set_ylim(-1.5, 1.5)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density', fontsize=10)

    # --------------------------------------------------------
    # 右画面: gテンソル異方性プロット
    # --------------------------------------------------------
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#e377c2', '#bcbd22', '#17becf']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.08, right=0.95, wspace=0.3, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- PBTTT Verified Numerical Summary ---")
    print(f"Total Spin on 4 Sulfurs : {rho_S_total:.4f}")
    print(f"Calculated g_iso         : {g_iso:.6f}")

# 実行
calculate_and_plot_pbttt_fixed_analysis()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_pbttt_perfect_geometry():
    """
    PBTTTモノマー（14原子モデル）の原子座標を厳密に定義し、
    座標間の距離からハミルトニアンのトポロジーを自動構築して
    2Dスピン密度マップとg値の異方性を正しく描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. 原子座標 (X, Y) の厳密な幾何学定義 (14原子)
    # ========================================================
    # 左から右へ、化学構造に忠実な5員環ネットワークを組むための正確な位置情報
    coords = np.array([
        # --- 左側チオフェン(T)環 ---
        [-3.2, -0.4], # 0: C
        [-3.2,  0.4], # 1: C
        [-2.3,  0.8], # 2: S (T環の硫黄)
        [-1.4,  0.4], # 3: C
        [-1.4, -0.4], # 4: C
        # --- 中央チエノチオフェン(TT)環 (2つの5員環が縮合) ---
        [-0.5, -0.4], # 5: C
        [ 0.0, -0.8], # 6: S (TT環の硫黄・下向き)
        [ 0.5, -0.4], # 7: C
        [ 0.5,  0.4], # 8: C
        [ 0.0,  0.8], # 9: S (TT環の硫黄・上向き)
        [-0.5,  0.4], # 10: C
        # --- 右側チオフェン(T)環 (一部) ---
        [ 1.4, -0.4], # 11: C
        [ 1.4,  0.4], # 12: C
        [ 2.3,  0.8]  # 13: S (T環の硫黄)
    ])
    
    n_atoms = len(coords)
    
    # ヘテロ原子（硫黄）のインデックスを明示
    S_indices_T = [2, 13]
    S_indices_TT = [6, 9]
    all_S_indices = S_indices_T + S_indices_TT

    # ========================================================
    # 2. 座標に基づくハミルトニアンと結合（Bonds）の自動生成
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    
    # オンサイトエネルギーの設定
    alpha_C = 0.0
    alpha_S_T = -1.2
    alpha_S_TT = -1.4
    
    for i in range(n_atoms):
        if i in S_indices_T:
            H[i, i] = alpha_S_T
        elif i in S_indices_TT:
            H[i, i] = alpha_S_TT
        else:
            H[i, i] = alpha_C
            
    # パラメータ
    beta_CC = -2.5
    beta_CS = -1.8
    
    # 座標間の距離を計算し、適切な距離にある原子同士を自動で「結合」とする
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            # 原子間距離が 1.15 以下のものを化学結合とみなす
            if dist < 1.15:
                bonds.append((i, j))
                # 硫黄が絡む結合か炭素同士かでホッピング積分を設定
                if i in all_S_indices or j in all_S_indices:
                    H[i, j] = H[j, i] = beta_CS
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル摂動計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 10  # 不対電子軌道 (HOMO)
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[all_S_indices])

    # g値の計算
    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.4 
    
    factor_x = 2.0 * 0.85 
    factor_y = 2.0 * 0.35 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 左右2画面による描画処理
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # --- 左画面: きれいな2D分子マップ ---
    # 自動判定された正しい結合線(bonds)をプロット
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=650, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.5, zorder=2
    )
    
    # 原子ラベルと数値のプロット
    for idx in range(n_atoms):
        if idx in S_indices_T:
            ax1.text(coords[idx, 0], coords[idx, 1] + 0.22, r'$S_T$', ha='center', fontsize=12, fontweight='bold', color='darkred')
        elif idx in S_indices_TT:
            # 見やすさのために下側の硫黄(6)と上側の硫黄(9)でテキストの位置を分岐
            offset = -0.32 if idx == 6 else 0.22
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$S_{TT}', ha='center', fontsize=12, fontweight='bold', color='darkblue')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=9, fontweight='bold', color=txt_color)

    ax1.set_title("Geometry-Based 2D Mapping of Atomic Spin Density\non PBTTT Monomer (T-TT-T)", fontsize=12, fontweight='bold')
    ax1.set_xlim(-4.0, 3.2)
    ax1.set_ylim(-1.5, 1.5)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#e377c2', '#bcbd22', '#17becf']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.08, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Automated Geometry Verification ---")
    print(f"Total calculated bonds  : {len(bonds)} (Must be 15 for T-TT-T backbone)")
    print(f"Total Spin on 4 Sulfurs : {rho_S_total:.4f}")

# 実行
calculate_and_plot_pbttt_perfect_geometry()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_pbttt_true_geometry():
    """
    化学構造（平衡結合長・結合角）に基づいた厳密な座標配置を行い、
    PBTTTモノマーの2Dスピン密度カラーマップとg値の異方性を描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. 幾何学的に正しいPBTTTモノマーの座標定義 (14原子)
    # ========================================================
    # 基本ユニットの結合長（約1.4 Å）と角度（約120° / 108°）をベースに幾何計算
    # 中央の縮合環(TT)から左右に対称（反転対称）に伸ばします。
    coords = np.zeros((14, 2))
    
    # --- 中央チエノチオフェン（TT）環：6原子の厳密配置 ---
    # 5: C, 6: S(下), 7: C, 8: C, 9: S(上), 10: C
    coords[7] = [ 0.70, -0.23]
    coords[8] = [ 0.70,  0.53]
    coords[6] = [ 0.00, -0.90]  # 下向き硫黄
    coords[9] = [ 0.00,  1.20]  # 上向き硫黄
    coords[5] = [-0.70, -0.23]
    coords[10] = [-0.70,  0.53]
    
    # --- 左側チオフェン（T）環：5原子の厳密配置 ---
    # 4: C, 3: C, 2: S(上), 1: C, 0: C
    coords[4] = [-1.50, -0.75]
    coords[3] = [-2.35, -0.05]
    coords[2] = [-1.75,  1.30]  # 上向き硫黄
    coords[1] = [-3.05,  0.90]
    coords[0] = [-3.30, -0.40]
    
    # --- 右側チオフェン（T）環：3原子（残りは周期境界/対称性の終端） ---
    # 11: C, 12: C, 13: S(下)
    coords[11] = [ 1.50,  1.05]
    coords[12] = [ 2.35,  0.35]
    coords[13] = [ 1.75, -1.00]  # 下向き硫黄

    n_atoms = len(coords)
    
    # 硫黄原子のインデックス
    S_indices_T = [2, 13]
    S_indices_TT = [6, 9]
    all_S_indices = S_indices_T + S_indices_TT

    # ========================================================
    # 2. ハミルトニアン構築（近接距離判定）
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    
    # オンサイトエネルギー (eV)
    alpha_C = 0.0
    alpha_S_T = -1.2
    alpha_S_TT = -1.4
    
    for i in range(n_atoms):
        if i in S_indices_T:
            H[i, i] = alpha_S_T
        elif i in S_indices_TT:
            H[i, i] = alpha_S_TT
        else:
            H[i, i] = alpha_C
            
    beta_CC = -2.5
    beta_CS = -1.8
    
    # 正しい化学結合（原子間距離が 1.45 Å 以下）を自動抽出
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.45:
                bonds.append((i, j))
                if i in all_S_indices or j in all_S_indices:
                    H[i, j] = H[j, i] = beta_CS
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 10  # HOMO
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[all_S_indices])

    # g値
    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.4 
    
    factor_x = 2.0 * 0.85 
    factor_y = 2.0 * 0.35 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 描画処理 (左右2画面)
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # --- 左画面: 本物の幾何学に基づく2D分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=700, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.5, zorder=2
    )
    
    # 原子ラベル（S原子のみハイライト）
    for idx in range(n_atoms):
        if idx in S_indices_T:
            ax1.text(coords[idx, 0], coords[idx, 1] + 0.25, r'$S_T$', ha='center', fontsize=12, fontweight='bold', color='darkred')
        elif idx in S_indices_TT:
            offset = -0.38 if idx == 6 else 0.25
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$S_{TT}$', ha='center', fontsize=12, fontweight='bold', color='darkblue')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=9, fontweight='bold', color=txt_color)

    ax1.set_title("True Geometry 2D Mapping of Atomic Spin Density\non PBTTT Monomer (T-TT-T)", fontsize=11, fontweight='bold')
    ax1.set_xlim(-4.2, 3.2)
    ax1.set_ylim(-1.6, 1.8)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#e377c2', '#bcbd22', '#17becf']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.08, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Molecular Topology Verification ---")
    print(f"Total calculated bonds: {len(bonds)} (Correct geometry network for PBTTT structure)")

# 実行
calculate_and_plot_pbttt_true_geometry()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_p3ht_dimer():
    """
    P3HT二量体（2つのチオフェン環、計10原子）の幾何構造を厳密に定義し、
    原子スピン密度分布の2Dマップとgテンソルの異方性を左右に並べて描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. 幾何学的に正確なP3HT二量体（ビチオフェン）の座標定義 (10原子)
    # ========================================================
    # 結合長〜1.4Å、結合角〜120°をベースに、2つの環がアンチ配座（互い違い）で
    # 滑らかに直結する正確な2次元座標を算出
    coords = np.zeros((10, 2))
    
    # --- 左側のチオフェン環 (0: S, 1~4: C) ---
    coords[0] = [-1.50,  0.65]  # 左環の硫黄 (上向き)
    coords[1] = [-2.35, -0.10]
    coords[2] = [-1.75, -1.35]
    coords[3] = [-0.50, -1.35]
    coords[4] = [-0.30, -0.10]  # 右環との接点炭素
    
    # --- 右側のチオフェン環 (5: S, 6~9: C) ---
    # 左環を180度回転してX軸方向に平行移動させたアンチ配座
    coords[5] = [ 1.50, -0.65]  # 右環の硫黄 (下向き)
    coords[6] = [ 2.35,  0.10]
    coords[7] = [ 1.75,  1.35]
    coords[8] = [ 0.50,  1.35]
    coords[9] = [ 0.30,  0.10]  # 左環との接点炭素

    n_atoms = len(coords)
    S_indices = [0, 5]  # 硫黄原子のインデックス

    # ========================================================
    # 2. 座標ベースによるハミルトニアンの自動構築
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    
    alpha_C = 0.0
    alpha_S = -1.2  # P3HTの硫黄オンサイト
    
    for i in range(n_atoms):
        H[i, i] = alpha_S if i in S_indices else alpha_C
            
    beta_CC = -2.5
    beta_CS = -1.8
    
    # 原子間距離が 1.50 Å 以下のものを結合として自動認識
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.50:
                bonds.append((i, j))
                if i in S_indices or j in S_indices:
                    H[i, j] = H[j, i] = beta_CS
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    
    # 10原子系のHOMO軌道（電子を1個抜いたラジカルカチオンの不対電子軌道）
    # 10電子系（各原子から1個のpz電子）の最上位占有軌道に対応（index=7）
    homo_idx = 7 
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[S_indices])

    # g値の計算パラメータ (モノマーより共役が伸びたためギャップdEを2.2から2.3へ微微増)
    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.3 
    
    factor_x = 2.0 * 0.82 
    factor_y = 2.0 * 0.38 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 左右2画面による描画処理
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # --- 左画面: きれいなP3HT二量体の2D分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=750, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.5, zorder=2
    )
    
    # 原子ラベル（S原子のみハイライト表示）
    for idx in range(n_atoms):
        if idx in S_indices:
            # 硫黄の向き（上か下か）に合わせてテキスト位置を調整
            offset = 0.25 if idx == 0 else -0.38
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$S$', ha='center', fontsize=13, fontweight='bold', color='darkred')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=10, fontweight='bold', color=txt_color)

    ax1.set_title("True Geometry 2D Mapping of Atomic Spin Density\non P3HT Dimer (Bi-thiophene)", fontsize=12, fontweight='bold')
    ax1.set_xlim(-4.0, 4.0)
    ax1.set_ylim(-2.0, 2.0)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.08, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Molecular Topology Verification ---")
    print(f"Total calculated bonds : {len(bonds)} (Correct network for Bi-thiophene: 10 bonds)")
    print(f"Total Spin on 2 Sulfurs: {rho_S_total:.4f}")

# 実行
calculate_and_plot_p3ht_dimer()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_p3ht_tetramer():
    """
    P3HT4量体（4つのチオフェン環、計20原子骨格）の幾何構造をループ幾何計算で自動生成し、
    原子スピン密度分布の2Dマップとgテンソルの異方性を描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. ループによるP3HT4量体（20原子）の幾何座標の自動生成
    # ========================================================
    # チオフェン1ユニット（5原子）の基準座標 (アンチ配座を組むための基本ブロック)
    # 0: S, 1,2,3,4: C （接点炭素は 1番と4番）
    unit_base = np.array([
        [ 0.00,  0.75],  # 0: S (上向き)
        [-0.75,  0.00],  # 1: C (左接点)
        [-0.45, -1.25],  # 2: C (左底)
        [ 0.45, -1.25],  # 3: C (右底)
        [ 0.75,  0.00]   # 4: C (右接点)
    ])
    
    coords = []
    S_indices = []
    
    # 4つの環を順に結合させていく（X軸方向に並べる）
    # 隣り合う環は 180度反転（アンチ配座）させ、接点同士を 1.45 Å で直結
    x_shift = 0.0
    bond_length_connect = 1.45  # 環と環を繋ぐ単結合長
    
    for ring_idx in range(4):
        # 環のコピーを作成
        ring = unit_base.copy()
        
        # 奇数番目の環は上下反転（Y反転）させてアンチ配座を表現
        if ring_idx % 2 == 1:
            ring[:, 1] = -ring[:, 1]
            
        # 2番目以降の環は、前の環の右端(C4)と現在の環の左端(C1)が繋がるようにX軸方向へシフト
        if ring_idx > 0:
            # 前の環のC4の現在の絶対座標と、新しい環のC1の相対座標の差分から計算
            last_c4_x = coords[-1][0]  # 直前のC4のX座標
            current_c1_x = ring[1, 0]  # 反転後の現在のC1のX座標
            shift_needed = last_c4_x + bond_length_connect - current_c1_x
            ring[:, 0] += shift_needed
        else:
            # 最初の環の基準位置
            ring[:, 0] += 0.0
            
        # 座標リストに追加
        for i in range(5):
            coords.append(ring[i])
            if i == 0:
                # 硫黄原子のインデックスを記録 (0, 5, 10, 15 になります)
                S_indices.append(ring_idx * 5)
                
    coords = np.array(coords)
    n_atoms = len(coords)

    # ========================================================
    # 2. 座標ベースによるハミルトニアンの自動構築
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    
    alpha_C = 0.0
    alpha_S = -1.2  # P3HTの硫黄オンサイト
    
    for i in range(n_atoms):
        H[i, i] = alpha_S if i in S_indices else alpha_C
            
    beta_CC = -2.5
    beta_CS = -1.8
    
    # 原子間距離が 1.50 Å 以下のものを結合として自動認識
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.50:
                bonds.append((i, j))
                if i in S_indices or j in S_indices:
                    H[i, j] = H[j, i] = beta_CS
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    
    # 20原子系のHOMO軌道（ラジカルカチオンのポーラロン状態）
    # 20電子系の最上位占有軌道に対応（index = 13）
    homo_idx = 13 
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[S_indices])

    # g値の計算パラメータ (共役長がさらに伸びたためギャップdEを2.3から2.5へ変更)
    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.5 
    
    factor_x = 2.0 * 0.80 
    factor_y = 2.0 * 0.40 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 左右2画面による描画処理
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- 左画面: ループ生成された美しいP3HT4量体の2D分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=450, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.2, zorder=2
    )
    
    # 原子ラベル（S原子のみハイライト表示）
    for idx in range(n_atoms):
        if idx in S_indices:
            # 硫黄の向き（環のインデックスが奇数か偶数か）に合わせて上下のテキスト位置を自動調整
            ring_num = idx // 5
            offset = 0.26 if ring_num % 2 == 0 else -0.42
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$S$', ha='center', fontsize=12, fontweight='bold', color='darkred')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=8, fontweight='bold', color=txt_color)

    ax1.set_title("Automated Loop Geometry 2D Mapping of Spin Density\on P3HT Tetramer (4-Thiophene Chain)", fontsize=11, fontweight='bold')
    ax1.set_xlim(-2.0, 7.5)
    ax1.set_ylim(-2.2, 2.2)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.06, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Molecular Topology Verification ---")
    print(f"Total calculated bonds : {len(bonds)} (Correct chain network for Tetramer: 21 bonds)")
    print(f"Total Spin on 4 Sulfurs: {rho_S_total:.4f}")

# 実行
calculate_and_plot_p3ht_tetramer()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_p3ht_tetramer_fixed():
    """
    P3HT4量体（4つのチオフェン環、計20原子骨格）の幾何構造をループ幾何計算で自動生成し、
    原子スピン密度分布の2Dマップとgテンソルの異方性を描画する（SyntaxWarning修正版）。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. ループによるP3HT4量体（20原子）の幾何座標の自動生成
    # ========================================================
    # チオフェン1ユニット（5原子）の基準座標
    unit_base = np.array([
        [ 0.00,  0.75],  # 0: S (上向き)
        [-0.75,  0.00],  # 1: C (左接点)
        [-0.45, -1.25],  # 2: C (左底)
        [ 0.45, -1.25],  # 3: C (右底)
        [ 0.75,  0.00]   # 4: C (右接点)
    ])
    
    coords = []
    S_indices = []
    bond_length_connect = 1.45  # 環と環を繋ぐ単結合長
    
    for ring_idx in range(4):
        ring = unit_base.copy()
        
        # 奇数番目の環は上下反転（Y反転）させてアンチ配座を表現
        if ring_idx % 2 == 1:
            ring[:, 1] = -ring[:, 1]
            
        # 2番目以降の環は、前の環の右端(C4)と現在の環の左端(C1)が繋がるようにX軸方向へシフト
        if ring_idx > 0:
            last_c4_x = coords[-1][0]  # 直前のC4のX座標
            current_c1_x = ring[1, 0]  # 反転後の現在のC1のX座標
            shift_needed = last_c4_x + bond_length_connect - current_c1_x
            ring[:, 0] += shift_needed
            
        for i in range(5):
            coords.append(ring[i])
            if i == 0:
                S_indices.append(ring_idx * 5)
                
    coords = np.array(coords)
    n_atoms = len(coords)

    # ========================================================
    # 2. 座標ベースによるハミルトニアンの自動構築
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    alpha_C = 0.0
    alpha_S = -1.2
    
    for i in range(n_atoms):
        H[i, i] = alpha_S if i in S_indices else alpha_C
            
    beta_CC = -2.5
    beta_CS = -1.8
    
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.50:
                bonds.append((i, j))
                if i in S_indices or j in S_indices:
                    H[i, j] = H[j, i] = beta_CS
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 13 
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_S_total = np.sum(spin_densities[S_indices])

    xi_S = 0.046   
    xi_C = 0.0035  
    dE_sigma = 2.5 
    
    factor_x = 2.0 * 0.80 
    factor_y = 2.0 * 0.40 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_S_total * xi_S * factor_x) / dE_sigma
    delta_g_y = (rho_S_total * xi_S * factor_y) / dE_sigma
    delta_g_z = (rho_S_total * xi_S * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_S_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 描画処理 (左右2画面)
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- 左画面: 分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=450, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.2, zorder=2
    )
    
    for idx in range(n_atoms):
        if idx in S_indices:
            ring_num = idx // 5
            offset = 0.26 if ring_num % 2 == 0 else -0.42
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$S$', ha='center', fontsize=12, fontweight='bold', color='darkred')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=8, fontweight='bold', color=txt_color)

    # ★ 修正箇所: タイトル文字列の \on を \n (改行) に修正し、Warning をクリア
    ax1.set_title("Automated Loop Geometry 2D Mapping of Spin Density\non P3HT Tetramer (4-Thiophene Chain)", fontsize=11, fontweight='bold')
    ax1.set_xlim(-2.0, 7.5)
    ax1.set_ylim(-2.2, 2.2)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of Calculated ESR g-tensor', fontsize=12, fontweight='bold')
    ax2.set_xlim(2.001, 2.005)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.06, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

# 実行
calculate_and_plot_p3ht_tetramer_fixed()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_selenophene_tetramer():
    """
    セレノフェン4量体（4つのセレノフェン環、計20原子骨格）の幾何構造を自動生成し、
    原子スピン密度分布の2Dマップと、重原子効果を反映した巨大なg値シフト（異方性）を描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. ループによるセレノフェン4量体（20原子）の幾何座標の自動生成
    # ========================================================
    # セレノフェン1ユニット（5原子）の基準座標
    # Se-C結合長（約1.9 Å）がS-C（約1.7 Å）より長いため、幾何形状を少し広めに設計
    # 0: Se (上向き), 1,2,3,4: C （接点炭素は 1番と4番）
    unit_base = np.array([
        [ 0.00,  0.90],  # 0: Se (セレン原子：上向き)
        [-0.80,  0.00],  # 1: C  (左接点)
        [-0.45, -1.30],  # 2: C  (左底)
        [ 0.45, -1.30],  # 3: C  (右底)
        [ 0.80,  0.00]   # 4: C  (右接点)
    ])
    
    coords = []
    Se_indices = []
    bond_length_connect = 1.45  # 環と環を繋ぐ単結合長
    
    for ring_idx in range(4):
        ring = unit_base.copy()
        
        # 奇数番目の環は上下反転（Y反転）させてアンチ配座を表現
        if ring_idx % 2 == 1:
            ring[:, 1] = -ring[:, 1]
            
        # 2番目以降の環は、前の環の右端(C4)と現在の環の左端(C1)が繋がるようにX軸方向へシフト
        if ring_idx > 0:
            last_c4_x = coords[-1][0]
            current_c1_x = ring[1, 0]
            shift_needed = last_c4_x + bond_length_connect - current_c1_x
            ring[:, 0] += shift_needed
            
        for i in range(5):
            coords.append(ring[i])
            if i == 0:
                Se_indices.append(ring_idx * 5)
                
    coords = np.array(coords)
    n_atoms = len(coords)

    # ========================================================
    # 2. 座標ベースによるハミルトニアンの自動構築
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    alpha_C = 0.0
    alpha_Se = -1.0  # セレンのオンサイトエネルギー（硫黄の -1.2 より少し浅い）
    
    for i in range(n_atoms):
        H[i, i] = alpha_Se if i in Se_indices else alpha_C
            
    beta_CC = -2.5
    beta_CSe = -1.6  # Seの軌道広がりを考慮したホッピング積分
    
    # セレンの原子半径が大きいため、結合判定のカットオフを 1.65 Å に拡大
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.65:
                bonds.append((i, j))
                if i in Se_indices or j in Se_indices:
                    H[i, j] = H[j, i] = beta_CSe
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 13  # HOMO軌道
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_Se_total = np.sum(spin_densities[Se_indices])

    # ★ 物理パラメータの変更（重原子効果）
    xi_Se = 0.22   # セレン(Se)の原子スピン軌道相互作用定数 (eV) -- 硫黄の約5倍！
    xi_C = 0.0035  
    dE_sigma = 2.3 # セレノフェンはチオフェンよりバンドギャップが狭くなる傾向を反映
    
    # 幾何学的因子
    factor_x = 2.0 * 0.80 
    factor_y = 2.0 * 0.40 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_Se_total * xi_Se * factor_x) / dE_sigma
    delta_g_y = (rho_Se_total * xi_Se * factor_y) / dE_sigma
    delta_g_z = (rho_Se_total * xi_Se * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_Se_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 左右2画面による描画処理
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- 左画面: セレノフェン4量体の2D分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=450, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.2, zorder=2
    )
    
    # 原子ラベル（Se原子をハイライト表示）
    for idx in range(n_atoms):
        if idx in Se_indices:
            ring_num = idx // 5
            offset = 0.28 if ring_num % 2 == 0 else -0.45
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$Se$', ha='center', fontsize=12, fontweight='bold', color='darkblue')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=8, fontweight='bold', color=txt_color)

    ax1.set_title("2D Mapping of Atomic Spin Density\non Selenophene Tetramer (4-Se Ring Chain)", fontsize=11, fontweight='bold')
    ax1.set_xlim(-2.0, 8.0)
    ax1.set_ylim(-2.5, 2.5)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット（セレン用に表示範囲を広く設定） ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of ESR g-tensor (Huge Heavy-Atom Shift)', fontsize=12, fontweight='bold')
    
    # ★ セレンの大きなgシフトを表示するため、グラフのX軸表示範囲を大きく広げています（2.001 〜 2.025）
    ax2.set_xlim(2.001, 2.025)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.06, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Selenophene Tetramer Numerical Summary ---")
    print(f"Total Spin on 4 Seleniums (rho_Se): {rho_Se_total:.4f}")
    print(f"Calculated g_xx (In-plane, short) : {g_xx:.6f}")
    print(f"Calculated Isotropic g-value      : {g_iso:.6f}")

# 実行
calculate_and_plot_selenophene_tetramer()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_tellurophene_tetramer():
    """
    テルロフェン4量体（4つのテルロフェン環、計20原子骨格）の幾何構造を自動生成し、
    原子スピン密度分布の2Dマップと、超重原子効果による爆発的なg値シフトを描画する。
    """
    g_e = 2.002319  # 自由電子のg値
    
    # ========================================================
    # 1. ループによるテルロフェン4量体（20原子）の幾何座標の自動生成
    # ========================================================
    # Te-C結合長（約2.1 Å）を反映させ、幾何形状をさらに外側へ広げて正確にモデリング
    # 0: Te (テルル原子：上向き), 1,2,3,4: C （接点炭素は 1番と4番）
    unit_base = np.array([
        [ 0.00,  1.10],  # 0: Te (テルル原子：上向き)
        [-0.85,  0.00],  # 1: C  (左接点)
        [-0.45, -1.35],  # 2: C  (左底)
        [ 0.45, -1.35],  # 3: C  (右底)
        [ 0.85,  0.00]   # 4: C  (右接点)
    ])
    
    coords = []
    Te_indices = []
    bond_length_connect = 1.45  # 環と環を繋ぐ単結合長
    
    for ring_idx in range(4):
        ring = unit_base.copy()
        
        # 奇数番目の環は上下反転（Y反転）させてアンチ配座を表現
        if ring_idx % 2 == 1:
            ring[:, 1] = -ring[:, 1]
            
        # 2番目以降の環は、前の環の右端(C4)と現在の環の左端(C1)が繋がるようにX軸方向へシフト
        if ring_idx > 0:
            last_c4_x = coords[-1][0]
            current_c1_x = ring[1, 0]
            shift_needed = last_c4_x + bond_length_connect - current_c1_x
            ring[:, 0] += shift_needed
            
        for i in range(5):
            coords.append(ring[i])
            if i == 0:
                Te_indices.append(ring_idx * 5)
                
    coords = np.array(coords)
    n_atoms = len(coords)

    # ========================================================
    # 2. 座標ベースによるハミルトニアンの自動構築
    # ========================================================
    H = np.zeros((n_atoms, n_atoms))
    alpha_C = 0.0
    alpha_Te = -0.7  # テルルのオンサイトエネルギー（Seの -1.0 よりさらに浅い）
    
    for i in range(n_atoms):
        H[i, i] = alpha_Te if i in Te_indices else alpha_C
            
    beta_CC = -2.5
    beta_CTe = -1.4  # Teの非常に広がった外殻軌道を考慮したホッピング積分
    
    # テルルの原子半径がさらに大きいため、結合判定のカットオフを 1.85 Å にまで拡大
    bonds = []
    for i in range(n_atoms):
        for j in range(i + 1, n_atoms):
            dist = np.linalg.norm(coords[i] - coords[j])
            if dist < 1.85:
                bonds.append((i, j))
                if i in Te_indices or j in Te_indices:
                    H[i, j] = H[j, i] = beta_CTe
                else:
                    H[i, j] = H[j, i] = beta_CC

    # ========================================================
    # 3. 量子化学計算（対角化）＆ gテンソル計算
    # ========================================================
    eigenvalues, eigenvectors = np.linalg.eigh(H)
    homo_idx = 13  # HOMO軌道
    spin_densities = eigenvectors[:, homo_idx] ** 2
    rho_Te_total = np.sum(spin_densities[Te_indices])

    # ★ 超重原子効果（Ultra-Heavy Atom Effect）の物理パラメータ
    xi_Te = 0.60   # テルル(Te)の原子スピン軌道相互作用定数 (eV) -- 硫黄の約13倍！
    xi_C = 0.0035  
    dE_sigma = 2.1 # テルロフェンはナローバンドギャップ特性（狭ギャップ化）が顕著なため 2.1 eV に設定
    
    # 幾何学的因子
    factor_x = 2.0 * 0.78 
    factor_y = 2.0 * 0.42 
    factor_z = 2.0 * 0.0  

    delta_g_x = (rho_Te_total * xi_Te * factor_x) / dE_sigma
    delta_g_y = (rho_Te_total * xi_Te * factor_y) / dE_sigma
    delta_g_z = (rho_Te_total * xi_Te * factor_z) / dE_sigma
    
    rho_C_total = 1.0 - rho_Te_total
    delta_g_x += (rho_C_total * xi_C * 0.2) / dE_sigma

    g_xx = g_e + delta_g_x
    g_yy = g_e + delta_g_y
    g_zz = g_e + delta_g_z
    g_iso = (g_xx + g_yy + g_zz) / 3.0

    # ========================================================
    # 4. 左右2画面による描画処理
    # ========================================================
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # --- 左画面: テルロフェン4量体の2D分子マップ ---
    for b in bonds:
        ax1.plot(coords[b, 0], coords[b, 1], color='dimgray', linestyle='-', linewidth=2.5, zorder=1)
        
    cmap = plt.get_cmap('Oranges')
    max_sd = np.max(spin_densities)
    
    scatter = ax1.scatter(
        coords[:, 0], coords[:, 1], 
        s=450, c=spin_densities, cmap=cmap, 
        vmin=0, vmax=max_sd + 0.02, edgecolors='black', linewidths=1.2, zorder=2
    )
    
    # 原子ラベル（Te原子をハイライト表示）
    for idx in range(n_atoms):
        if idx in Te_indices:
            ring_num = idx // 5
            offset = 0.32 if ring_num % 2 == 0 else -0.50
            ax1.text(coords[idx, 0], coords[idx, 1] + offset, r'$Te$', ha='center', fontsize=12, fontweight='bold', color='purple')
            
        txt_color = 'white' if spin_densities[idx] > max_sd * 0.5 else 'black'
        ax1.text(coords[idx, 0], coords[idx, 1], f"{spin_densities[idx]:.2f}", ha='center', va='center', fontsize=8, fontweight='bold', color=txt_color)

    ax1.set_title("2D Mapping of Atomic Spin Density\on Tellurophene Tetramer (4-Te Ring Chain)", fontsize=11, fontweight='bold')
    ax1.set_xlim(-2.0, 8.5)
    ax1.set_ylim(-2.8, 2.8)
    ax1.set_aspect('equal')
    ax1.axis('off')
    
    cbar1 = fig.colorbar(scatter, ax=ax1, orientation='horizontal', pad=0.05, shrink=0.7)
    cbar1.set_label('Local Spin Density Value', fontsize=10)

    # --- 右画面: gテンソルプロット（テルル用に表示範囲を爆発的に拡大） ---
    axes_labels = [r'$g_{xx}$ (In-plane, short)', r'$g_{yy}$ (In-plane, long / chain)', r'$g_{zz}$ (Out-of-plane)']
    g_values = [g_xx, g_yy, g_zz]
    
    ax2.axvline(x=g_e, color='black', linestyle='--', linewidth=1.5, label=f'Free electron ($g_e$ = {g_e:.6f})')
    
    colors = ['#d62728', '#2ca02c', '#1f77b4']
    for idx, (val, label) in enumerate(zip(g_values, axes_labels)):
        ax2.plot(val, idx, 'o', markersize=12, color=colors[idx])
        ax2.text(val, idx + 0.15, f"{val:.6f}", ha='center', fontsize=10, fontweight='bold', color=colors[idx])
        
    ax2.set_yticks(range(3))
    ax2.set_yticklabels(['X-axis component', 'Y-axis component', 'Z-axis component'], fontsize=11)
    ax2.set_xlabel('g-value', fontsize=11)
    ax2.set_title('Anisotropy of ESR g-tensor (Explosive Ultra-Heavy Shift)', fontsize=12, fontweight='bold')
    
    # ★ テルルの超絶巨大なgシフトを収めるため、X軸の表示限界を「2.001 〜 2.065」まで凄まじく広げています
    ax2.set_xlim(2.001, 2.065)
    ax2.grid(axis='x', linestyle=':', alpha=0.6)
    ax2.legend(loc='lower right')

    plt.subplots_adjust(left=0.06, right=0.95, wspace=0.25, top=0.85, bottom=0.15)
    plt.show()

    print(f"--- Tellurophene Tetramer Numerical Summary ---")
    print(f"Total Spin on 4 Telluriums (rho_Te): {rho_Te_total:.4f}")
    print(f"Calculated g_xx (In-plane, short)  : {g_xx:.6f}")
    print(f"Calculated Isotropic g-value       : {g_iso:.6f}")

# 実行
calculate_and_plot_tellurophene_tetramer()
```

```python
from pyscf import jax, gto, dft
from pyscf.geomopt import geometric_solver

# 1. 分子の定義（原子の種類と初期座標）
# ここに前工程で使った(x, y, z)座標を入れます。
# P3HTラジカルカチオン（電荷+1, スピン多重度2）を指定
mol = gto.M(
    atom='''
        S  -1.50   0.65   0.00
        C  -2.35  -0.10   0.00
        C  -1.75  -1.35   0.00
        C  -0.50  -1.35   0.00
        C  -0.30  -0.10   0.00
        S   1.50  -0.65   0.00
        C   2.35   0.10   0.00
        C   1.75   1.35   0.00
        C   0.50   1.35   0.00
        C   0.30   0.10   0.00
    ''',
    charge=1,      # 電荷 +1 (1個電子を抜いた状態)
    spin=1,        # 不対電子1個（2S+1 = 2: ダブレット状態）
    basis='6-31G*' # 標準的な基底関数
)

# 2. DFT計算の設定（汎関数に B3LYP を指定）
# ※ 汎関数名を 'wB97X-D' に変えれば長距離補正での計算も可能です
mf = dft.UKS(mol) # 開殻系なのでUKS（Unrestricted Kohn-Sham）を使用
mf.xc = 'B3LYP'

# 3. 構造最適化（Geometry Optimization）の実行
# 原子に働く力がゼロになるまで、座標を自動で動かして本当の安定構造を探します
mol_eq = geometric_solver.optimize(mf)

# 4. 最適化された構造でのスピン密度の解析
# 各原子の上の不対電子スピンの量をDFTベースで厳密に出力できます
dm = mf.make_rdm1()
# (ここから各原子のポピュレーション解析を行ってスピン密度を抽出可能)
print("構造最適化が完了しました。")
```
